```python
from transformers import pipeline
summarizer = pipeline("summarization")
text = """Artificial intelligence (AI) has revolutionized industries across the globe.
From healthcare to finance, AI is automating processes, improving decision-making,
and unlocking new possibilities.However, as AI continues to advance, it brings
challenges related to ethics, data privacy, and job displacement."""
summary = summarizer(text, max_length=50, min_length=25,
do_sample=False)
print(summary)
```

```python
from google.colab import files

# Step 1: Load the summarization pipeline
summarizer = pipeline("summarization")

# Step 2: Upload your file
print("Please upload a text file:")
uploaded = files.upload()  # This opens a file upload dialog in Colab

# Step 3: Process the uploaded file
for filename in uploaded.keys():
    print(f"Processing file: {filename}")
    with open(filename, 'r') as file:
        text = file.read()

    # Step 4: Summarize the content
    summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
    print(f"Summary for {filename}:\n{summary[0]['summary_text']}")
```
